# Linear Regression

## General equation of straight line

$$y = b0*x+b1$$

## Slope of the straight line

$$b0 = \frac{\sum{(x-\bar{x})(y-\bar{y})}}{\sum{(x-\bar{x})^{2}}}$$

## Straight line intercept

$$b1 = \bar{y} - b0 \cdot \bar{x}$$


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def estimate_b0_b1(x, y):
    n = np.size(x)

    # average
    m_x, m_y = np.mean(x), np.mean(y)

    # calculate sum of xy and sum xx
    sum_xy = np.sum((x-m_x)*(y-m_y))
    sum_xx = np.sum(x*(x-m_x))
    print(np.sum(x*(x-m_x)) == np.sum((x-m_x)**2))

    # regression
    b_1 = sum_xy / sum_xx
    b_0 = m_x - b_1 * m_x

    return (b_0, b_1)
```

```python
def plot_regression(x, y, b):
    plt.scatter(x, y, color = "g", marker = "o", s = 30)

    # vector of predictions
    y_pred = b[0] + b[1]*x
    plt.plot(x, y_pred, color = "b")

    # labels
    plt.xlabel('X')
    plt.ylabel('Y')

    plt.show()
```

```python
def main():
    x = np.array([1,2,3,4,5])
    y = np.array([1,4,3,4,5])

    b = estimate_b0_b1(x, y)
    print(f'b0 = {b[0]}, b1 = {b[1]}')

    # graph
    plot_regression(x, y, b)

if __name__ == "__main__":
    main()
```
